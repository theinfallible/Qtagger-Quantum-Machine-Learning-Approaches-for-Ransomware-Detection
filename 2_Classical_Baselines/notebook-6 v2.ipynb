```python
# --- Step 1: Install XGBoost ---
# This cell should be run first if XGBoost is not already installed.
!pip install -q xgboost

import numpy as np
import os
import time
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import subprocess

# Import scikit-learn and XGBoost components
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Ignore warnings for cleaner output
warnings.filterwarnings('ignore')

def load_and_prep_data(file_path: str, is_multiclass: bool, test_size: float = 0.2, subsample_size: int = None):
    """
    Loads data, optionally subsamples it, handles binary vs. multi-class logic, splits, and scales.
    """
    print("\n" + "="*60)
    print(f"LOADING DATA FOR: {os.path.basename(file_path)}")
    print(f"Task Type: {'Multi-Class' if is_multiclass else 'Binary'}")
    print("="*60)

    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    data = np.load(file_path)
    X = data[:, :-1]
    y = data[:, -1].astype(int)
    
    print(f"Original data shape: {data.shape}")
    
    # --- NEW: Subsampling Logic ---
    if subsample_size is not None and subsample_size < len(X):
        print(f"--- Subsampling data to {subsample_size} total samples ---")
        # Use train_test_split to perform a stratified subsample
        X, _, y, _ = train_test_split(
            X, y, train_size=subsample_size, random_state=42, stratify=y
        )
        print(f"Subsampled data shape: {X.shape}")

    n_features = X.shape[1]
    unique_labels, counts = np.unique(y, return_counts=True)
    n_classes = len(unique_labels)
    print(f"Found {n_classes} unique classes: {unique_labels}")
    print(f"Class distribution: {dict(zip(unique_labels, counts))}")

    # --- Data Handling Logic ---
    if not is_multiclass:
        if n_classes > 2:
            print(f"WARNING: Data has {n_classes} classes but is marked for binary classification.")
            print("Mapping Class 0 -> Benign (0), Classes 1+ -> Ransomware (1)")
        
        y_final = np.where(y == 0, 0, 1)
        X_final = X
        
        print("Balancing the binary dataset...")
        class_0_indices = np.where(y_final == 0)[0]
        class_1_indices = np.where(y_final == 1)[0]
        min_samples = min(len(class_0_indices), len(class_1_indices))
        
        balanced_0 = np.random.choice(class_0_indices, min_samples, replace=False)
        balanced_1 = np.random.choice(class_1_indices, min_samples, replace=False)
        
        balanced_indices = np.concatenate([balanced_0, balanced_1])
        np.random.shuffle(balanced_indices)
        
        X_final = X_final[balanced_indices]
        y_final = y_final[balanced_indices]

    else:
        X_final, y_final = X, y

    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X_final, y_final, test_size=test_size, random_state=42, stratify=y_final
    )

    # Scale the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    print("\n--- DATASET SUMMARY ---")
    print(f"Number of features: {n_features}")
    print(f"Training samples: {len(X_train)}")
    print(f"Testing samples: {len(X_test)}")

    return X_train_scaled, X_test_scaled, y_train, y_test
    
def is_gpu_available():
    """Checks if an NVIDIA GPU is available and accessible."""
    try:
        subprocess.check_output('nvidia-smi')
        return True
    except (subprocess.CalledProcessError, FileNotFoundError):
        return False

def train_and_evaluate_model(model_name: str, X_train, X_test, y_train, y_test):
    """
    Initializes, trains, and provides in-depth evaluation for a specified classical model.
    """
    print("\n" + "-"*50)
    print(f"PERFORMANCE ANALYSIS FOR: {model_name.upper()}")
    print("-"*50)

    # Initialize the chosen model with stronger default parameters
    if model_name.lower() == 'randomforest':
        model = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1, class_weight='balanced')
    
    elif model_name.lower() == 'xgboost':
        if is_gpu_available():
            print("INFO: NVIDIA GPU detected. Configuring XGBoost for GPU acceleration.")
            model = XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.1, 
                                  random_state=42, use_label_encoder=False, 
                                  eval_metric='mlogloss', tree_method='gpu_hist')
        else:
            print("INFO: No GPU detected. Configuring XGBoost for CPU.")
            model = XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.1,
                                  random_state=42, use_label_encoder=False, 
                                  eval_metric='mlogloss', n_jobs=-1)
    else:
        raise ValueError("Model not supported. Choose 'RandomForest' or 'XGBoost'.")

    # Train the model and time the process
    print(f"\nTraining {model_name}...")
    start_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - start_time
    print(f"Training complete in {training_time:.2f} seconds.")

    # Make predictions
    print("\nEvaluating on the test set...")
    y_pred = model.predict(X_test)

    # Determine class labels for reporting
    n_classes = len(np.unique(np.concatenate([y_train, y_test])))
    if n_classes == 2:
        target_names = ["Benign (Class 0)", "Ransomware (Class 1)"]
    else:
        target_names = [f"Class {i}" for i in range(n_classes)]
    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nOverall Test Set Accuracy: {accuracy:.4f}\n")
    
    print("--- Detailed Classification Report ---")
    print(classification_report(y_test, y_pred, target_names=target_names, digits=4, zero_division=0))

    # Plot confusion matrix
    print("\n--- Confusion Matrix ---")
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10 if n_classes > 5 else 7, 8 if n_classes > 5 else 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
    plt.title(f'{model_name} - Confusion Matrix', fontsize=16)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.show()

# ==============================================================================
# --- MAIN EXECUTION: AUTOMATION CONTROL PANEL ---
# ==============================================================================
if __name__ == "__main__":
    
    # --- 1. CONFIGURE YOUR SUBSAMPLE SIZE HERE ---
    # Set to an integer (e.g., 500) to test with a small sample.
    # Set to `None` to use the full dataset.
    SUBSAMPLE_SIZE = 500
    
    # --- 2. CONFIGURE YOUR EXPERIMENTS HERE ---
    experiments = [
        # --- Binary Classification Tasks (is_multiclass = False) ---
        {'file_path': "/kaggle/input/mydataset/RANSAP_win7-120gb-hdd_2CLASS.npy", 'is_multiclass': False},
        {'file_path': "/kaggle/input/mydataset/RANSAP_win7-120gb-ssd_2CLASS.npy", 'is_multiclass': False},

        # --- Multi-Class Classification Tasks (is_multiclass = True) ---
        {'file_path': "/kaggle/input/mydataset/RANSAP_win7-120gb-hdd_12CLASS.npy", 'is_multiclass': True},
        {'file_path': "/kaggle/input/mydataset/RANSAP_win7-120gb-ssd_12CLASS.npy", 'is_multiclass': True},
        
        # --- 26-Class Tasks (is_multiclass = True) ---
        {'file_path': "/kaggle/input/mydataset/RANSAP_win7-120gb-hdd_26CLASS.npy", 'is_multiclass': True},
        {'file_path': "/kaggle/input/mydataset/RANSAP_win7-250gb-hdd_26CLASS.npy", 'is_multiclass': True},
    ]
    
    models_to_run = ['RandomForest', 'XGBoost']
    
    # --- AUTOMATION LOOP ---
    for experiment in experiments:
        file = experiment['file_path']
        
        if not os.path.exists(file):
            print(f"\nWARNING: File not found, skipping experiment -> '{file}'")
            continue
            
        # Load data for the current experiment, passing the subsample size
        X_train, X_test, y_train, y_test = load_and_prep_data(
            file_path=file,
            is_multiclass=experiment['is_multiclass'],
            subsample_size=SUBSAMPLE_SIZE
        )
        
        # Run both classical models on the loaded data
        for model_name in models_to_run:
            train_and_evaluate_model(
                model_name=model_name,
                X_train=X_train, X_test=X_test,
                y_train=y_train, y_test=y_test
            )
            
    print("\n" + "="*60)
    print("ALL BENCHMARKING EXPERIMENTS COMPLETE!")
    print("="*60)
```
